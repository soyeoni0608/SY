## 6. Ïó∞ÎèÑ, ÏõîÎ≥ÑÎ°ú Í∑∏Î£πÌôîÌïòÏó¨ Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑùÌïòÍ∏∞

### (1-1) CA Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ïó∞ÎèÑ, ÏõîÎ≥ÑÎ°ú Í∑∏Î£πÌôîÌïòÏó¨ Îç∞Ïù¥ÌÑ∞Î•º Î∂ÑÏÑùÌïòÍ∏∞ ÏúÑÌï¥ publish_time Î≥ÄÏàòÏóêÏÑú yearÏôÄ month Î∂ÑÎ¶¨

```python
def Convert(string): 
    li = list(string.split("-")) 
    return li 

CA_df = pd.DataFrame(CA['publish_time'])
CA_publish_month=[]
CA_publish_year=[]
for i in range(len(CA)):
    time = CA_df.loc[i,'publish_time']
    time_list = Convert(time)
    CA_publish_year.append(time_list[0])
    CA_publish_month.append(time_list[1])
```

### (1-2) DE Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ïó∞ÎèÑ, ÏõîÎ≥ÑÎ°ú Í∑∏Î£πÌôîÌïòÏó¨ Îç∞Ïù¥ÌÑ∞Î•º Î∂ÑÏÑùÌïòÍ∏∞ ÏúÑÌï¥ publish_time Î≥ÄÏàòÏóêÏÑú yearÏôÄ month Î∂ÑÎ¶¨

```python
DE_df = pd.DataFrame(DE['publish_time'])
DE_publish_month=[]
DE_publish_year=[]
for i in range(len(DE)):
    time = DE_df.loc[i,'publish_time']
    time_list = Convert(time)
    DE_publish_year.append(time_list[0])
    DE_publish_month.append(time_list[1])
```

### (2-1) CAÏóêÏÑú yearÏôÄ monthÎ•º DataFrameÏúºÎ°ú Î≥ÄÌôò

```python
CA_publish_year_df=pd.DataFrame(CA_publish_year, columns=['year'])
CA_publish_month_df=pd.DataFrame(CA_publish_month, columns=['month'])
```
